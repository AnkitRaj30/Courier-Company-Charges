<h1><center>Calculated the expected charges based on the zones & Type of Shipment for company X </center></h1>

- **The charges are dependent upon two factors:**
  - Weight of the product
  - Distance between the warehouse (pickup location) and customer’s delivery address (destination location)


- **forward charges =fwd charges**
  - 0.5kg = fwd_fixed 
  - 0.5kg + each*(0.5kg)=fwd_fixed+(additional_weight * fwd_additional)

- **forward & rto charges = fwd+rto**
  - 0.5kg = fwd_fixed + fwd_rto
  - 0.5kg + each*(0.5kg) = fwd_fixed+fwd_rto + (additional_weight * (fwd_additional + rto_additional))

```python
Expected_charge_as_per_X=[]
for i in range(len(Order_Invoice)):
    fwd_category='fwd_'+Order_Invoice['Zone (X)'][i]
    fwd_fixed=Courier_Rates[fwd_category+'_fixed'][0]
    fwd_additional=Courier_Rates[fwd_category+'_additional'][0]
    rto_category='rto_'+Order_Invoice['Zone (X)'][i]
    rto_fixed=Courier_Rates[rto_category+'_fixed'][0]
    rto_additional=Courier_Rates[rto_category+'_additional'][0]
    if Order_Invoice['Type of Shipment'][i]=='Forward charges':
        if Order_Invoice['Weight_slab_X (KG)'][i]<=0.5:
            Expected_charge_as_per_X.append(fwd_fixed)
        else:
            additional_weight=(Order_Invoice['Weight_slab_X (KG)'][i]-0.5)/0.5
            Expected_charge_as_per_X.append(fwd_fixed+additional_weight*fwd_additional)
    elif Order_Invoice['Type of Shipment'][i]=='Forward and RTO charges':
        if Order_Invoice['Weight_slab_X (KG)'][i]<=0.5:
            Expected_charge_as_per_X.append(fwd_fixed+rto_fixed)
        else:
            additional_weight=(Order_Invoice['Weight_slab_X (KG)'][i]-0.5)/0.5
            Expected_charge_as_per_X.append(fwd_fixed+rto_fixed+additional_weight*(fwd_additional+rto_additional))

Order_Invoice['Expected Charge As Per x (Rs.)']= Expected_charge_as_per_X 
```

- **Created the three Column for Charge type , count & amount**
   
   - **Charge type :-**  
     - "Under_Charged     = Courier company is charging less from company x as compared company x expected charge"
     - "Correctly_charged = Courier company is charging correct from company x as compared company x expected charge
     - "Over_Charged"     = Courier company is charging more from company x as compared company x expected charge 
   
   - **Count       :- Count the how many order under,over & correctly charged**
   - **Amount      :- Sum of the total Amount**

```python
Charge_Type =[]
Count=[]
Amount=[]
```

- **Based On Under_Charge**
  - **Count :- Calculated Under_Charged Order**
  - **Amount:- Sum Of the Values from " Difference Between Expected Charges and Billed Charges (Rs.) " column for Checking total Profit**

```python
under_charged=Order_invoice[Order_invoice['Difference Between Expected Charges and Billed Charges (Rs.)'] > 0.0]
under=["Total Orders where x has been under_charged",len(under_charged),sum(under_charged['Difference Between Expected Charges and Billed Charges (Rs.)'])]
Charge_Type.append(under[0])
Count.append(under[1])
Amount.append(under[2])
```

- **Based On Correct_Charge**
  - **Count :- Calculated Correct_Charged Order**
  - **Amount:- Sum Of the Values from " Billing Amount By CC (Rs.) " column for Checking total Correct Charge**

```python
correctly_charged=Order_invoice[Order_invoice['Difference Between Expected Charges and Billed Charges (Rs.)'] == 0.0]
correctly=["Total Orders where x has been correctly_charged",len(correctly_charged),sum(correctly_charged['Billing Amount By CC (Rs.)'])]
Charge_Type.append(correctly[0])
Count.append(correctly[1])
Amount.append(correctly[2])
```

- **Based On Over_Charge**
  - **Count :- Calculated Over_Charged Order**
  - **Amount:- Sum Of the amount from " Difference Between Expected Charges and Billed Charges (Rs.) " column for Checking total Loss**

```python
over_charged=Order_invoice[Order_invoice['Difference Between Expected Charges and Billed Charges (Rs.)'] < 0.0]
over=["Total Orders where x has been over_charged",len(over_charged),sum(over_charged['Difference Between Expected Charges and Billed Charges (Rs.)'])]
Charge_Type.append(over[0])
Count.append(over[1])
Amount.append(over[2])
```

- **Created a new Dataframe(table) with Three column : Charge_Type,Count & Amount**

```python
Total_Order = pd.DataFrame()
```

```python
Total_Order["Charge_Type"]=Charge_Type
Total_Order["Count"]=Count
Total_Order["Amount"]=Amount
```
